# Values For Graph

```python
# map data for all four graphs from weather data frame
latitude = weather_data_df["Lat"]
max_temp = weather_data_df["Max Temp"]
humidity = weather_data_df["Humidity"]
cloudiness = weather_data_df["Cloudiness"]
wind_speed = weather_data_df["Wind Speed"]
```

# Latitude vs Temperature Scattered Graph

```python
# First scatter graph city latitude vs max temperature
plt.figure(figsize=(8,6))
plt.scatter(latitude,
           max_temp,
           edgecolor='k',
           linewidths=1,
           marker="o",
            alpha=1,
           label = "Cities")
# Set graph labels, titles
plt.title("City Latitude vs Max Temperature (2018-06-13)")
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True, color='w', linestyle='-', linewidth=1.25)
plt.gca().patch.set_facecolor('0.91')
plt.savefig("Resources/Fig1")
plt.show()
```

# Latitude vs Humidity Scattered Graph

```python
# First scatter graph city latitude vs humidity
plt.figure(figsize=(8,6))
plt.scatter(latitude,
           humidity,
           edgecolor="black",
           linewidths=1,
           marker="o",
            alpha=1,
           label = "Cities")
# Set graph labels, titles
plt.title("City Latitude vs Humidity (2018-06-13)")
plt.ylabel("Humidity (%)")
plt.xlabel("Latitude")
plt.grid(True, color='w', linestyle='-', linewidth=1.25)
plt.gca().patch.set_facecolor('0.91')
plt.savefig("Resources/Fig2")
plt.show()
```
